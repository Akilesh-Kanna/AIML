```python
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Life Expectancy Data.csv')

# Rename columns to remove trailing spaces
data.columns = data.columns.str.strip()

# Remove rows with missing values in the relevant columns
data = data.dropna(subset=['Adult Mortality', 'BMI', 'HIV/AIDS', 'Income composition of resources', 'Schooling', 'Life expectancy'])

# Define the independent variables (X) and dependent variable (Y)
X = data[['Adult Mortality', 'BMI', 'HIV/AIDS', 'Income composition of resources', 'Schooling']]
Y = data['Life expectancy']

# Add a constant term to the independent variables (for intercept)
X_with_const = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X_with_const).fit()

# Get the predicted values
predictions = model.predict(X_with_const)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y, predictions)

# Print the summary of the model
print(model.summary())

# Print R-squared and MSE
print(f"R-squared: {model.rsquared:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Life Expectancy Data.csv')

# Rename columns to remove trailing spaces
data.columns = data.columns.str.strip()

# Remove rows with missing values in the relevant columns
data = data.dropna(subset=['Adult Mortality', 'BMI', 'HIV/AIDS', 'Income composition of resources', 'Schooling', 'Life expectancy'])

# Define the independent variables (X) and dependent variable (Y)
X = data[['Adult Mortality', 'BMI', 'HIV/AIDS', 'Income composition of resources', 'Schooling']]
Y = data['Life expectancy']

# Add a constant term to the independent variables (for intercept)
X_with_const = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X_with_const).fit()

# Get the predicted values
predictions = model.predict(X_with_const)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y, predictions)

# Create the plot for Adult Mortality vs Life Expectancy
plt.figure(figsize=(10, 6))

# Plot Adult Mortality
plt.subplot(2, 3, 1)
plt.scatter(data['Adult Mortality'], Y, color='blue', label='Actual Data', alpha=0.6)
plt.plot(data['Adult Mortality'], predictions, color='red', label='Regression Line')
plt.title('Adult Mortality vs Life Expectancy')
plt.xlabel('Adult Mortality')
plt.ylabel('Life Expectancy')
plt.legend()

# Plot BMI
plt.subplot(2, 3, 2)
plt.scatter(data['BMI'], Y, color='blue', label='Actual Data', alpha=0.6)
plt.plot(data['BMI'], predictions, color='red', label='Regression Line')
plt.title('BMI vs Life Expectancy')
plt.xlabel('BMI')
plt.ylabel('Life Expectancy')
plt.legend()

# Plot HIV/AIDS
plt.subplot(2, 3, 3)
plt.scatter(data['HIV/AIDS'], Y, color='blue', label='Actual Data', alpha=0.6)
plt.plot(data['HIV/AIDS'], predictions, color='red', label='Regression Line')
plt.title('HIV/AIDS vs Life Expectancy')
plt.xlabel('HIV/AIDS')
plt.ylabel('Life Expectancy')
plt.legend()

# Plot Income Composition
plt.subplot(2, 3, 4)
plt.scatter(data['Income composition of resources'], Y, color='blue', label='Actual Data', alpha=0.6)
plt.plot(data['Income composition of resources'], predictions, color='red', label='Regression Line')
plt.title('Income Composition vs Life Expectancy')
plt.xlabel('Income Composition')
plt.ylabel('Life Expectancy')
plt.legend()

# Plot Schooling
plt.subplot(2, 3, 5)
plt.scatter(data['Schooling'], Y, color='blue', label='Actual Data', alpha=0.6)
plt.plot(data['Schooling'], predictions, color='red', label='Regression Line')
plt.title('Schooling vs Life Expectancy')
plt.xlabel('Schooling')
plt.ylabel('Life Expectancy')
plt.legend()

# Display the R-squared and MSE on the plot
plt.subplot(2, 3, 6)
plt.axis('off')
plt.text(0.1, 0.9, f'R-squared: {model.rsquared:.4f}\nMSE: {mse:.4f}', fontsize=12)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
