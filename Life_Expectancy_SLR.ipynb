```python
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Life Expectancy Data.csv')

# Rename columns to remove trailing spaces
data.columns = data.columns.str.strip()

# Remove rows with missing values in the relevant columns
data = data.dropna(subset=['Adult Mortality', 'Life expectancy'])

# Define the independent (X) and dependent (Y) variables
X = data['Adult Mortality']
Y = data['Life expectancy']

# Add a constant term to the independent variable (for intercept)
X_with_const = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X_with_const).fit()

# Get the predicted values
predictions = model.predict(X_with_const)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y, predictions)

# Print R-squared and MSE
print(f"R-squared: {model.rsquared:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Life Expectancy Data.csv')

# Rename columns to remove trailing spaces
data.columns = data.columns.str.strip()

# Remove rows with missing values in the relevant columns
data = data.dropna(subset=['Adult Mortality', 'Life expectancy'])

# Define the independent (X) and dependent (Y) variables
X = data['Adult Mortality']
Y = data['Life expectancy']

# Add a constant term to the independent variable (for intercept)
X_with_const = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X_with_const).fit()

# Get the predicted values
predictions = model.predict(X_with_const)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y, predictions)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual Data', alpha=0.6)
plt.plot(X, predictions, color='red', label='Regression Line')

# Add labels, title, and legend
plt.title('Simple Linear Regression: Life Expectancy vs Adult Mortality', fontsize=14)
plt.xlabel('Adult Mortality', fontsize=12)
plt.ylabel('Life Expectancy', fontsize=12)
plt.legend()

# Display R-squared and MSE on the plot
plt.text(0.05, 0.95, f'R-squared: {model.rsquared:.4f}\nMSE: {mse:.4f}',
         transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

# Show the plot
plt.grid(True)
plt.show()
```
